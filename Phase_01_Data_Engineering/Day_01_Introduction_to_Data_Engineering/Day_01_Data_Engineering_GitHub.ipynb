```python
import pandas as pd

# Load dataset
df = pd.read_csv("student_performance.csv")

# Display first 5 rows
df.head()
```

```python
print("=" * 50)
print("         STUDENT DATA EXPLORER")
print("=" * 50)
```

```python
print("Dataset Overview")
print("-" * 30)

print(f"Total number of students : {len(df)}")
print(f"Total number of columns  : {len(df.columns)}")
print(f"Unique departments       : {df['department'].nunique()}")
```

```python
print("Department Wise Student Count")
print("-" * 35)

print(df["department"].value_counts())
```

```python
print("Score Analysis")
print("-" * 30)

print(f"Highest Math Score            : {df['math_score'].max()}")
print(f"Lowest Math Score             : {df['math_score'].min()}")
print(f"Average Math Score            : {df['math_score'].mean():.2f}")
print(f"Average Programming Score     : {df['programming_score'].mean():.2f}")
print(f"Average Attendance Percentage : {df['attendance_percentage'].mean():.2f}%")
```

```python
df['total_score']=df['math_score']+df['science_score']+df['english_score']+df['programming_score']
print("Average Total Score by Gender")
print("-" * 40)

print(df.groupby("gender")["total_score"].mean())
```

```python
print("Top 5 Performers")
print("-" * 30)

top_students = df.sort_values(by="total_score", ascending=False).head(5)

print(top_students[["name", "department", "total_score"]])
```

```python
print("Students with Attendance Below 75%")
print("-" * 45)

low_attendance = df[df["attendance_percentage"] < 75]

if len(low_attendance) > 0:
    print(low_attendance[["name", "department", "attendance_percentage"]])
else:
    print("All students have attendance above 75%")
```

Practice Questions

```python
electronics_students = df[df["department"] == "Electronics"]

print("Number of students from Electronics Department:", len(electronics_students))
```

```python
female_attendance = df[df["gender"] == "Female"]["attendance_percentage"].mean()

print(f"Average attendance of female students: {female_attendance:.2f}%")
```

```python
lowest_programming = df.loc[df["programming_score"].idxmin()]

print("Student with Lowest Programming Score")
print("-" * 40)

print(f"Name : {lowest_programming['name']}")
print(f"Programming Score : {lowest_programming['programming_score']}")
```

```python
students_above_80 = df[df["math_score"] > 80]

print("Number of students scored above 80 in math:", len(students_above_80))
```
